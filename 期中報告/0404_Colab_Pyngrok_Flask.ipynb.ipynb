```python
from google.colab import userdata

ngrok_authtoken = userdata.get('NGROK_AUTHTOKEN')  #從 Colab Secrets 取得預先存好的 ngrok 驗證金鑰
port=5051   # 設定 Flask 程式要對應的連接埠 (Port)
```

```python
import os
from pyngrok import ngrok
```

```python
ngrok.kill()  # 先清除現有的 ngrok 連線，避免重複佔用
```

```python
ngrok.set_auth_token(ngrok_authtoken)    # 設定 ngrok 權杖進行身分驗證
tunnel = ngrok.connect(port, name="linebot_tunnel")   # 建立一個隧道 (Tunnel)，將內部的 port 5051 映射到一個隨機產生的公開網址
print(tunnel.public_url)     # 輸出公開網址
```

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "<h1>B12090036 辛育齊 連線成功</h1>"      # 當使用者造訪網址時，回傳這段 HTML

if __name__ == '__main__':
    app.run(port=port)
```
